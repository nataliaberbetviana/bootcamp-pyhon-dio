```python
# Estruturas condicionais
saldo = 500
saque = 200
limite = 1000

if saldo >= saque and saque <= limite: # Verifica se o saldo é suficiente e se o saque está dentro do limite
    print('Saque realizado com sucesso!') # Realiza o saque
    saldo -= saque # Atualiza o saldo
    print(f'Seu saldo atual é {saldo} reais.') # Mostra o saldo atualizado
else:
    print('Saque não realizado. Verifique o saldo e o limite.') # Exibe mensagem de erro
    print(f'Seu saldo atual é {saldo} reais.') # Mostra o saldo atual

idade = int(input('Digite sua idade: ')) # Converte a entrada para inteiro
if idade < 18: # Verifica se a idade é menor que 18
    print('Você é menor de idade.')
elif idade == 18: # Verifica se a idade é igual a 18
    print('Você acabou de se tornar maior de idade.')
else: # Caso contrário
    print('Você é maior de idade.') 
```

```python
# Estrturas de repetição
# Laço while
contador = 1
while contador <= 10: # Enquanto o contador for menor ou igual a 10
    print(contador) # Imprime o valor do contador
    contador += 1 # Incrementa o contador em 1
print('Fim do laço while!')

# Laço for
for numero in range(1, 11): # Itera sobre uma sequência de números de
    print(numero) # Imprime o valor do número
print('Fim do laço for!')

# Laço for com lista
lista = [1, 2, 3, 4, 5]
for item in lista: # Itera sobre os elementos da lista
    print(item) # Imprime o valor do item 
print('Fim do laço for!')

# Laço for com string
for letra in 'Python': # Itera sobre os caracteres da string
    print(letra) # Imprime o valor da letra
print('Fim do laço for!')

# Laço for com dicionário
pessoa = {'nome': 'Guilherme', 'idade': 23, 'cidade': 'São Paulo'}
for chave in pessoa: # Itera sobre as chaves do dicionário
    print(chave, pessoa[chave]) # Imprime a chave e o valor correspondente
print('Fim do laço for!')   

# Laço for com tupla
tupla = (1, 2, 3, 4, 5)
for item in tupla: # Itera sobre os elementos da tupla
    print(item) # Imprime o valor do item
print('Fim do laço for!')

# Laço for com set
conjunto = {1, 2, 3, 4, 5}
for item in conjunto: # Itera sobre os elementos do conjunto
    print(item) # Imprime o valor do item
print('Fim do laço for!')

# Laço for com break e continue
for numero in range(1, 11): # Itera sobre uma sequência de números de 1 a 10
    if numero == 5: # Se o número for igual a 5
        print('Número 5 encontrado, saindo do laço.') 
        break # Sai do laço
    if numero % 2 == 0: # Se o número for par
        continue # Pula para a próxima iteração
    print(numero) # Imprime o valor do número
print('Fim do laço for!')

# Laço while com break e continue
contador = 1
while contador <= 10: # Enquanto o contador for menor ou igual a 10
    if contador == 5: # Se o contador for igual a 5
        print('Contador 5 encontrado, saindo do laço.')
        break # Sai do laço
    if contador % 2 == 0: # Se o contador for par
        contador += 1 # Incrementa o contador em 1
        continue # Pula para a próxima iteração
    print(contador) # Imprime o valor do contador
    contador += 1 # Incrementa o contador em 1
print('Fim do laço while!')
```

```python
# Manipulação de Strings e Fatiamento
texto = 'Curso de Python'
print(texto)  # Imprime a string completa
print(texto[0])  # Imprime o primeiro caractere
print(texto[5])  # Imprime o sexto caractere
print(texto[-1])  # Imprime o último caractere
print(texto[-6])  # Imprime o sexto caractere a partir do final
print(texto[0:5])  # Imprime os caracteres do índice 0 ao 4
print(texto[5:])  # Imprime os caracteres do índice 5 até o final
print(texto[:5])  # Imprime os caracteres do início até o índice 4
print(texto[0:10:2])  # Imprime os caracteres do índice 0 ao 9, pulando de 2 em 2
print(texto[::-1])  # Imprime a string invertida
print(len(texto))  # Imprime o tamanho da string
print(texto.lower())  # Converte a string para minúsculas
print(texto.upper())  # Converte a string para maiúsculas
print(texto.split())  # Divide a string em uma lista de palavras
print(' '.join(texto))  # Junta os caracteres da string com espaço entre eles
print(texto.replace('Python', 'Java'))  # Substitui 'Python' por 'Java'
print(texto.find('Python'))  # Encontra o índice onde 'Python' começa
print(texto.count('o'))  # Conta quantas vezes 'o' aparece na string
print(texto.startswith('Curso'))  # Verifica se a string começa com 'Curso'
print(texto.endswith('Python'))  # Verifica se a string termina com 'Python'
print(texto.strip())  # Remove espaços em branco no início e no fim da string
print(texto.isalpha())  # Verifica se todos os caracteres são letras
print(texto.isdigit())  # Verifica se todos os caracteres são dígitos
print(texto.isalnum())  # Verifica se todos os caracteres são alfanuméricos
print(texto.isspace())  # Verifica se todos os caracteres são espaços em branco
print(f'O texto é: {texto}')  # Imprime a string usando f-strings
print(texto.title())  # Converte a string para o formato título
print(texto.center(30, '-'))  # Centraliza a string em um campo de 30 caracteres, preenchendo com '-'
print(texto.center(50, '*'))  # Centraliza a string em um campo de 80 caracteres, preenchendo com '*'
print(texto.ljust(30, '-'))  # Alinha a string à esquerda em um campo de 30 caracteres, preenchendo com '-'
print(texto.rjust(30, '-'))  # Alinha a string à direita em um campo de 30 caracteres, preenchendo com '-'
print(texto.zfill(30))  # Preenche a string com zeros à esquerda até atingir 30 caracteres
```

```python
# Interpolação de variáveis
nome = 'Guilherme'
idade = 23
altura = 1.75
peso = 70.5
imc = peso / (altura ** 2)
print(f'O {nome} tem {idade} anos, {altura} de altura e pesa {peso} kg.')
print(f'O IMC de {nome} é {imc:.2f}.')  # Formata o IMC com 2 casas decimais
print('O {} tem {} anos, {} de altura e pesa {} kg.'.format(nome, idade, altura, peso))
print('O {0} tem {1} anos, {2} de altura e pesa {3} kg.'.format(nome, idade, altura, peso))
print('O {n} tem {i} anos, {a} de altura e pesa {p} kg.'.format(n=nome, i=idade, a=altura, p=peso))
print('O %s tem %d anos, %.2f de altura e pesa %.2f kg.' % (nome, idade, altura, peso))  # Formata o peso com 1 casa decimal
```

```python
# Listas em Python
frutas = ['banana', 'maçã', 'laranja']
print(frutas)  # Imprime a lista completa
print(frutas[0])  # Imprime o primeiro elemento
print(frutas[1])  # Imprime o segundo elemento
print(frutas[-1])  # Imprime o último elemento
print(frutas[0:2])  # Imprime os dois primeiros elementos
print(frutas[1:])  # Imprime do segundo elemento até o final
print(frutas[:2])  # Imprime do início até o segundo elemento
print(frutas[::2])  # Imprime todos os elementos, pulando de 2 em 2
print(frutas[::-1])  # Imprime a lista invertida
print(len(frutas))  # Imprime o tamanho da lista
frutas.append('uva')  # Adiciona 'uva' ao final da lista
print(frutas)
frutas.insert(1, 'morango')  # Adiciona 'morango' na posição 1
print(frutas)
frutas.remove('maçã')  # Remove 'maçã' da lista
print(frutas)
fruta_removida = frutas.pop()  # Remove o último elemento e retorna ele
print(frutas)
print(fruta_removida)
fruta_removida = frutas.pop(1)  # Remove o elemento na posição 1 e retorna ele
print(frutas)
print(fruta_removida)
frutas[0] = 'abacaxi'  # Altera o primeiro elemento para 'abacaxi'
print(frutas)
frutas.sort()  # Ordena a lista em ordem alfabética
print(frutas)
frutas.sort(reverse=True)  # Ordena a lista em ordem alfabética reversa
print(frutas)
frutas.reverse()  # Inverte a ordem dos elementos na lista
print(frutas)
print(frutas.index('laranja'))  # Retorna o índice de 'laranja'
print(frutas.count('banana'))  # Conta quantas vezes 'banana' aparece na lista
frutas.clear()  # Remove todos os elementos da lista
print(frutas)
letras = list('Python')  # Cria uma lista a partir de uma string
print(letras)
frutas = ['banana', 'maçã', 'laranja']
lista = list(frutas[0]) # Cria uma lista a partir do primeiro elemento da lista frutas
print(lista)
```

```python
# Métodos úteis de listas
numeros = [5, 3, 8, 1, 2]
numeros2 = numeros.copy()  # Cria uma cópia da lista
print(numeros2)
print(min(numeros))  # Retorna o menor valor da lista
print(max(numeros))  # Retorna o maior valor da lista
print(sum(numeros))  # Retorna a soma dos valores da lista
print(sorted(numeros))  # Retorna uma nova lista ordenada
print(sorted(numeros, reverse=True))  # Retorna uma nova lista ordenada em ordem reversa
print(list(reversed(numeros)))  # Retorna uma nova lista com os elementos na ordem inversa
print(numeros)  # A lista original permanece inalterada
print(len(numeros))  # Retorna o tamanho da lista
print(numeros.extend([10, 20, 30]))  # Adiciona os elementos de outra lista ao final da lista
print(numeros)
numeros3 = [1, 2, 3, 4, 5, 2, 5, 6, 2, 6]
print(numeros3)
numero_procurado = 2
while numero_procurado in numeros3:  # Enquanto o número procurado estiver na lista
    numeros3.remove(numero_procurado)  # Remove o número procurado da lista
print(numeros3)  # Imprime a lista sem o número procurado
```

```python
# Tuplas em Python
tupla = (1, 2, 3, 4, 5)
print(tupla)  # Imprime a tupla completa
print(tupla[0])  # Imprime o primeiro elemento
print(tupla[1])  # Imprime o segundo elemento
print(tupla[-1])  # Imprime o último elemento
print(tupla[0:2])  # Imprime os dois primeiros elementos
print(tupla[1:])  # Imprime do segundo elemento até o final
print(tupla[:2])  # Imprime do início até o segundo elemento
print(tupla[::2])  # Imprime todos os elementos, pulando de 2 em 2
print(tupla[::-1])  # Imprime a tupla invertida
print(len(tupla))  # Imprime o tamanho da tupla
print(tupla.count(2))  # Conta quantas vezes o valor 2 aparece na tupla
print(tupla.index(3))  # Retorna o índice do valor 3
# tupla[0] = 10  # Isso gera um erro, pois tuplas são imutáveis
nova_tupla = tupla + (6, 7, 8)  # Cria uma nova tupla concatenando outra tupla
print(nova_tupla)
tupla2 = (9, 10)
tupla3 = tupla + tupla2  # Cria uma nova tupla concatenando outra tupla
print(tupla3)
tupla4 = tupla * 2  # Cria uma nova tupla repetindo os elementos
print(tupla4)
a, b, c, d, e = tupla  # Desempacotamento
print(a, b, c, d, e)
# a, b, c = tupla  # Isso gera um erro, pois o número de variáveis não corresponde ao número de elementos na tupla
a, b, *resto = tupla  # Desempacotamento com resto
print(a, b, resto)

# Métodos úteis de tuplas
tupla = (5, 3, 8, 1, 2)
print(tupla.count(3))  # Conta quantas vezes o valor 3 aparece na tupla
print(tupla.index(8))  # Retorna o índice do valor 8
print(len(tupla))  # Retorna o tamanho da tupla
print(min(tupla))  # Retorna o menor valor da tupla
print(max(tupla))  # Retorna o maior valor da tupla
print(sum(tupla))  # Retorna a soma dos valores da tupla
print(sorted(tupla))  # Retorna uma nova lista ordenada
print(sorted(tupla, reverse=True))  # Retorna uma nova lista ordenada em ordem reversa
print(tupla)  # A tupla original permanece inalterada
print(list(reversed(tupla)))  # Retorna uma nova lista com os elementos na ordem inversa
print(tupla)  # A tupla original permanece inalterada
print(tupla * 2)  # Cria uma nova tupla repetindo os elementos
```

```python
# Dicionários em Python
pessoa = {'nome': 'Guilherme', 'idade': 23, 'cidade': 'São Paulo'}
print(pessoa)  # Imprime o dicionário completo
print(pessoa['nome'])  # Imprime o valor da chave 'nome'
print(pessoa['idade'])  # Imprime o valor da chave 'idade'
print(pessoa.get('cidade'))  # Imprime o valor da chave 'cidade'
print(pessoa.get('pais', 'Brasil'))  # Imprime o valor da chave 'pais' ou 'Brasil' se a chave não existir
pessoa['idade'] = 24  # Altera o valor da chave 'idade'
print(pessoa)
pessoa['pais'] = 'Brasil'  # Adiciona uma nova chave 'pais
print(pessoa)
del pessoa['cidade']  # Remove a chave 'cidade'
print(pessoa)
pessoa.pop('pais')  # Remove a chave 'pais' e retorna o valor
print(pessoa)
chave_removida = pessoa.pop('nome')  # Remove a chave 'nome' e retorna o valor
print(pessoa)
print(chave_removida)
print(len(pessoa))  # Imprime o tamanho do dicionário
print(pessoa.keys())  # Imprime as chaves do dicionário
print(pessoa.values())  # Imprime os valores do dicionário
print(pessoa.items())  # Imprime os itens (chave, valor) do dicionário
for chave in pessoa:  # Itera sobre as chaves do dicionário
    print(chave, pessoa[chave])  # Imprime a chave e o valor correspondente
for chave, valor in pessoa.items():  # Itera sobre os itens (chave, valor) do dicionário
    print(chave, valor)  # Imprime a chave e o valor correspondente
```
